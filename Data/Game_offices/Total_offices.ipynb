```python


city_counts = df['City'].value_counts()
city_counts.plot(kind='bar', color='black', edgecolor='black')

plt.xticks(rotation=45, ha='right')
plt.title('Count of Companies in Each City')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
pastel_colors = sns.color_palette("pastel")

# Plot the pie chart
city_counts.plot(kind='pie', colors=pastel_colors, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Companies in Each City')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
